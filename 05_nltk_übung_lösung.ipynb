* Einlesen der Datei "DatenAnaonymisiert_ansi_part.csv"
* Eine Variable "daten_nltk" erstellen, die nur die Spalte "UNFALLHERGANG" enthält
* Eine Variable "daten_tokens" erstellen, die nur die erste Zeile des Objekts "daten_nltk" enthält und den Satz mit der Funktion nltk.word_tokenize tokenisiert
* Wie häufig kommt das Wort "Versicherter" im ersten Satz vor? Dazu die Funktion concordance verwenden
* An welchen Stellen kommen die Wörter "ist" und "Versicherter" vor? Dazu die Funktion dispersion_plot verwenden
* Erstellen Sie eine Liste über die Häufigkeit, mit der einzelne Tokens vorkommen und speichern diese in die Variable "counts"
* Lassen Sie sich die Liste in einer Grafik mit der Funktion "plot" visualisieren


```python
import pandas as pd
import nltk
```

```python
nltk.download('punkt')
```

```python
daten = pd.read_csv("/content/drive/MyDrive/vbg/DatenAnonymisiert_ansi_part.csv",encoding='latin1',delimiter=';', index_col=1)
```

```python
daten["UNFALLHERGANG"][1]
```

```python
daten_nltk = daten["UNFALLHERGANG"]
daten_nltk
```

```python
daten_tokens = nltk.word_tokenize(daten_nltk[1])
daten_tokens
```

```python
nltk.Text(daten_tokens).concordance('Versicherter')
```

```python
nltk.Text(daten_tokens).dispersion_plot(["ist", "Versicherter"])
```

```python
counts = nltk.FreqDist(daten_tokens)
counts
```

```python
counts.plot(20, cumulative=False)
```

```python
daten_nltk_tokens = daten_nltk.apply(nltk.word_tokenize)
daten_nltk_tokens
```

```python
counts_2 = daten_nltk_tokens.apply(nltk.FreqDist)
counts_2
```

```python
nltk.FreqDist(sum(daten_nltk_tokens, [])).plot(20, cumulative=False)
```

```python
tokenizer = nltk.RegexpTokenizer("[\w']+")
daten_nltk_tokens = daten_nltk.apply(tokenizer.tokenize)
daten_nltk_tokens
```

```python
nltk.FreqDist(sum(daten_nltk_tokens, [])).plot(20, cumulative=False)
```
