```python
daten_full = pd.read_csv("/content/drive/MyDrive/vbg/DatenAnonymisiert_ansi_part.csv",encoding='latin1',delimiter=';', index_col=1)
daten_full_grouped = daten_full.groupby('UNFART_BEZ')

daten_full_grouped

# iterate over each group
for group_name, df_group in daten_full_grouped:
  daten_selected = daten_full[daten_full['UNFART_BEZ'] == group_name]
  daten_selected = daten_selected["UNFALLHERGANG"]
  daten_selected =  daten_selected.str.replace(r'(\n)', r' ')
  np.savetxt('test_'+ group_name + '.txt', daten_selected.values,fmt='%s')
```

```python
daten_full = pd.read_csv("/content/drive/MyDrive/vbg/DatenAnonymisiert_ansi.CSV",encoding='latin1',delimiter=';', index_col=1)
daten_full_grouped = daten_full.groupby('UNFART_NAM')

daten_full_grouped

# iterate over each group
for group_name, df_group in daten_full_grouped:
  daten_selected = daten_full[daten_full['UNFART_NAM'] == group_name]
  daten_selected = daten_selected["UNFALLHERGANG"]
  daten_selected =  daten_selected.str.replace(r'(\n)', r' ')
  np.savetxt('/content/drive/MyDrive/vbg/test_'+ str(group_name) + '.txt', daten_selected.values,fmt='%s')
```

```python
import nltk
nltk.download('punkt')
filename=open("/content/drive/MyDrive/vbg/test_part.txt","r")
tokenizer = nltk.RegexpTokenizer("[\w']+")
tokens = []
for line in filename.readlines():
    tokens+=tokenizer.tokenize(line)
tokens = [token.lower() for token in tokens]
nltk.FreqDist(tokens).plot(20, cumulative=False)
```
